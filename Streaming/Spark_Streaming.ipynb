```python
#Simulate Continuous Sensor Data Stream (CSV)
import os
import time
import pandas as pd
import random
from datetime import datetime

stream_dir = "/content/sensor_stream"
os.makedirs(stream_dir, exist_ok=True)

sensor_ids = ['sensor_A', 'sensor_B', 'sensor_C']

# Generate and write data continuously for simulation
for i in range(10):  # Adjust range for longer streams
    rows = []
    for sensor in sensor_ids:
        rows.append({
            "timestamp": datetime.now().isoformat(),
            "sensor_id": sensor,
            "temperature": round(random.uniform(20.0, 35.0), 2),
            "humidity": round(random.uniform(30.0, 70.0), 2)
        })
    df = pd.DataFrame(rows)
    df.to_csv(f"{stream_dir}/sensor_data_{i}.csv", index=False)
    time.sleep(2)  # simulate delay between batches
```

```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("SensorStreamSimulation").getOrCreate()
spark.sparkContext.setLogLevel("ERROR")
```

```python
from pyspark.sql.types import StructType, StringType, DoubleType, TimestampType

file_schema = StructType() \
    .add("timestamp", TimestampType()) \
    .add("sensor_id", StringType()) \
    .add("temperature", DoubleType())\
    .add("humidity", DoubleType())
```

```python
from pyspark.sql.functions import col

# Define schema automatically from header and read CSVs as stream
df = spark.readStream.option("header", "true").schema(file_schema).csv(stream_dir)

# Filter and show average temperature by sensor
query = df.groupBy("sensor_id").avg("temperature").writeStream \
    .outputMode("complete") \
    .format("console") \
    .start()
query.awaitTermination()  # Run stream for 15 seconds
```

```python
from pyspark.streaming import StreamingContext
from pyspark import SparkContext

sc = SparkContext()

```

```python
ssc = StreamingContext(sc,1)
```

```python
lines = ssc.socketTextStream("rtd.hpwren.ucsd.edu", 12028)
```

```python
# Parse a line of weather station data, returning the average wind direction measurement
#
import re
def parse(line):
    print(line)
    match = re.search("(\d+)", line)
    if match:
        val = match.group(1)
        return [int(val)]
    return []
```

```python
vals = lines.flatMap(parse)
```

```python
window = vals.window(5,3)
```

```python
def stats(rdd):
    print(rdd.collect())
    if rdd.count() > 0:
        print("max = {}, min = {}".format(rdd.max(), rdd.min()))
```

```python
window.foreachRDD(lambda rdd: stats(rdd))
```
