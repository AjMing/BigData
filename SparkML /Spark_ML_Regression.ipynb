#  Regression

```python
from pyspark.sql import DataFrameNaFunctions
from pyspark.ml.feature import VectorAssembler
```

```python
df.count(), len(df.columns)
```

```python
df = df.drop('number')
```

```python
df = df.na.drop()
```

```python
df.count(), len(df.columns)
```

```python
df.columns
```

```python
featureColumns =df['air_pressure_9am',
   'air_temp_9am',
 'rain_accumulation_9am',
 'relative_humidity_9am']

featureColumns=featureColumns.columns
```

```python
featureColumns=df.columns[:-1]
```

```python
featureColumns
```

```python
assembler = VectorAssembler(inputCols=featureColumns, outputCol="features")
assembled = assembler.transform(df)
```

```python
assembled.show(10)
```

```python
select_assembled=assembled.select("relative_humidity_3pm","features")
select_assembled.show()
```

```python
select_assembled=select_assembled.withColumnRenamed("relative_humidity_3pm","label")
select_assembled.show()
```

```python
(trainData, testData) = select_assembled.randomSplit([0.8,0.2], seed = 13234 )
```

```python
trainData.count(),testData.count()
```

```python
from pyspark.ml.regression import LinearRegression
```

```python
trainData.show()
```

```python
testData.show()
```

```python
lr = LinearRegression(labelCol="relative_humidity_3pm")
```

```python
model1 = lr.fit(trainData)
#input= features column
#label = label column
```

```python
print("Coefficients: %s" % str(model1.coefficients))
print("Intercept: %s" % str(model1.intercept))
```

```python
trainingSummary = model1.summary
print("numIterations: %d" % trainingSummary.totalIterations)
print("objectiveHistory: %s" % str(trainingSummary.objectiveHistory))
trainingSummary.residuals.show()
print("RMSE: %f" % trainingSummary.rootMeanSquaredError)
print("r2: %f" % trainingSummary.r2)
```

```python
predictions = model1.transform(testData)
```

```python
predictions.show()
```

```python
from pyspark.ml.evaluation import RegressionEvaluator

Evaluator=RegressionEvaluator(labelCol="relative_humidity_3pm",
predictionCol="prediction", metricName="rmse")

rmse=Evaluator.evaluate(predictions)
```

```python
Evaluator=RegressionEvaluator(labelCol="relative_humidity_3pm",
predictionCol="prediction", metricName="r2")

r2=Evaluator.evaluate(predictions)
```

```python
print("RMSE: %f" % rmse)
print("r^2: %f" % r2)
```
